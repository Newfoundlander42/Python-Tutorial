# Function in Python:
A function in python is a named reusable block of code that can called multiple times in a program to execute things stored in it.

1. Before a function can be used in python it must defined as so: `def function_name():` and then it's constituent code is written in the next lines indented like an `if` block, the () here can be thought of as arguement storing vessels. Their meaning shall become clearer as so: <br>
   <br>
   
   ```
   def hello(target_name):
      print("Hello ", target_name)
   ``` 
<br> Here the function `hello` stores a parameter `target_name` defined by the user. <br>
   When this function is 'called' as so `hello(naman)` it understands that a function called `hello` is being called and `naman` is being initiated as an arguement.

The Python function `except` is used to deal with eceptional scenarios such as index out of bounds and print out a statement in case they arise.
The code <br>
```python
def same_chars(string, ing1, ing2):
    if ing1>=len(string) or ing2>=len(string):
        return False
    if string[ing1]==string[ing2]:
        return True
    else:
        return False
```
<br>
can also be used as so with the except command:

````python
def same_chars(string, ing1, ing2):
    try:
        if string[ing1]==string[ing2]:
            return True
        else:
            return False
    except IndexError:
        return False
if __name__ == "__main__":
    print(same_chars("coder", 1, 2))
    print(same_chars("emotional", 3, 10))
        
````

Here the `try` statement makes the code try to execute our code, in case though if it fails in the exceptional case of an Index Error, then the `except` statement makes it disregard the error and execute a secondary statement (here: `return False`)
