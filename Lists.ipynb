Here are an example code which utlizies various functions:



````python
def palindromes(str1):
    reverse=""
    for i in range ((len(str1)-1),-1,-1):
        reverse+=str1[i]
    if reverse==str1:
        return True
    else:
        return False
    
while True:
    inp=input("Please type in a palindrome:")
    if palindromes(inp)==True:
        print(f"{inp} is a palindrome!")
        break
    else:
        print("that wasn't a palindrome")
````

Another place where lists are employed within for loops are to find out the best or the worst among it based on some criteria, for example, here is a sample code to find all the longest words in a list and print all of them:

````python
def all_the_longest(my_list):
    best=[my_list[0]]
    for i in range(1, len(my_list)):
        if len(my_list[i])==len(best[0]):
            best.append(my_list[i])
        elif len(my_list[i])>len(best[0]):
            best.clear()
            best.append(my_list[i])
        else:
            continue
    return best
if __name__=="__main__":
    my_list = ["adele", "mark", "dorothy", "tim", "hedy", "richard"]
    result = all_the_longest(my_list)
    print(result)
````

Some bits of code showing various methods are:

````python
def everything_reversed(my_list):
    navin=[]
    
    for i in range(0,len(my_list)):
        revers=my_list[i]
        navin.append(revers[::-1])
    return navin[::-1]

if __name__=="__main__":
    my_list = ["Hi", "there", "example", "one more"]
    new_list = everything_reversed(my_list)
    print(new_list)
````

````python
def most_common_character(str1):
    best=str1[0]
    
    for i in range(1, len(str1)):
        if str1.count(str1[i])>str1.count(best):
            best =str1[i]
        else:
            continue
    return best

if __name__=="__main__":
    first_string = "abcdbde"
    print(most_common_character(first_string))
    
    second_string = "exemplaryelementary"
    print(most_common_character(second_string))
````

````python
def longest_series_of_neighbours(lattice_energy):
    best=1
    current=1
    for i in range (1, len(lattice_energy)):
        if abs(lattice_energy[i]-lattice_energy[i-1])==1:
            current+=1
            best=max(best, current)
        else:
            current=1
    return best

if __name__=="__main__":
    my_list = [1, 2, 5, 4, 3, 4]
    print(longest_series_of_neighbours(my_list))
````

## Working with Matrices 

2 Dimensionally matrices are essentially just a list of lists, i.e. a list inside a list, however to talk of 3 dimensional matrices we would go one step further by placing lists inside lists inside lists, so an $\mathbb{R}$ dimensional matrix is just $\mathbb{R}-1$ lists nested inside a super or main list.

To navigate through a Matrix we need two coordinates, the $x$ and $y$ also called the $i$ and $j$ components.
<br> For example the following matrix can be represented in Python as ```my_matrix = [[1, 2, 3], [3, 2, 1], [4, 5, 6]]```
<br> <br>   <br> <br>
Any element of this list can be accessed by using its address i.e. `my_matrix[i][j]` to call forth it.
>Note: To traverse through the elements of a matrix a `for` loop is used and to traverse through the column of a matrix we nest a `for loop` inside another with the given constraints.
