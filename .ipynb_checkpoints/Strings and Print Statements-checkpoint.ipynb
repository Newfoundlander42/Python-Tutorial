# Strings
Strings are an integral part of any piece of code, they are the link of communication between the user and the code. <br>
An important operator in strings is the `string_name[]` the character inside the square brackets calls the index of a letter. All strings are stored as indices in python <br> <br>
 <br>
The index addresses begin from 0 and go to the last character. <br>
To call the last character one can use `my_string[len(string)-1]` or simply use `my_string[-1]`, the last character is stored as -1 and the negative indexing goes on from there. <br>

> 1. Strings can be added and subtracted like integers and floats, the + operator combines strings whereas the * operator rewrites a string a given number of time without leaving a space between them.

An example code that uses these functions to frame a word is:

````python
word=input("Word: ") 
l=(27-len(word))//2
r=(28-len(word))//2
print("*"*30)
print(f"* {' ' * l}{word}{' ' * r}*")
print("*"*30)
````

### f-strings:
`f-strings` or Formatted literal strings are an amazing way of formatting strings not just in print commands but also in variables: 
- We can use f-strings to control the amount of decimals to be printed in a float point variable like so:
- ```number = 1/3``` <br> ```print(f"The number is {number:.2f}")``` <br>
- Here the clon character seperates the variable to printed and the format specifier `.2f`; the 2 here specifies that the result should be displayed utpo 2 decimals and as a float type number as indicated by the `f`. <br>
- The format specifier `:15` tells python to leave a whitespace of 15 after it, i.e. to right align it, whereas `:>15`  tells it to leave a whitespace of 15 before the variable, i.e. left align it.


````python
names =  [ "Steve", "Jean", "Katherine", "Paul" ]
for name in names:
  print(f"{name:15} centre {name:>15}")
````

````python
def formatted(lattice_energy):
    navin=[]
    for number in lattice_energy:
        navin.append(f"{number:.2f}")
    return navin

if __name__=="__main__":
    lattice_energy=[1.1111111, 2.3456789, 2, 3.5678]
    print(formatted(lattice_energy))
````

# The Art of writing Cleaner Code
As one starts getting more and more knowledge about various functions and methods, it is important to focus on not just writing code to make things work, but also writing it to be short, clean and understandable
Here is a piece of code that does the job of grading the user:

````python
def main():
    ep=[] #exam points
    exe=[] #exercises completed

    while True:

        marks=input("Exam points and exercises completed:")
        if marks=="":
            break

        intermediate=marks.split()
        ep.append(int(intermediate[0]))
        exe.append(int(intermediate[1]))

    return ep,exe

def analysis():
    ep, exe=main()
    courses=len(ep)
    np=[] #new points


    for i in range(0,len(exe)):
        if exe[i]<10:
            np.append(0)
        elif 10<=exe[i]<20:
            np.append(1)
        elif 20<=exe[i]<30:
            np.append(2)
        elif 30<=exe[i]<40:
            np.append(3)
        elif 40<=exe[i]<50:
            np.append(4)
        elif 50<=exe[i]<60:
            np.append(5)
        elif 60<=exe[i]<70:
            np.append(6)
        elif 70<=exe[i]<80:
            np.append(7)
        elif 80<=exe[i]<90:
            np.append(8)
        elif 90<=exe[i]<100:
            np.append(9)
        elif 100==exe[i]:
            np.append(10)
    grade1, grade2, grade3, grade4, grade5, fail=grade(np,ep)
    pavg=(sum(ep)+sum(np))/courses
    passper=((courses-fail)/courses)*100
    return pavg, passper, fail, grade1, grade2, grade3, grade4, grade5

def grade(l1, l2):
    grade1=0
    grade2=0
    grade3=0
    grade4=0
    grade5=0
    grade0=0
    fail=0
    for i in range(0, len(l1)):
        
        if l1[i]+l2[i]<=14 or l2[i]<10:
            fail+=1 #intermediate fail variable
        elif 15<=l1[i]+l2[i]<=17:
            grade1+=1
        elif 18<=l1[i]+l2[i]<=20:
            grade2+=1
        elif 21<=l1[i]+l2[i]<=23:
            grade3+=1
        elif 24<=l1[i]+l2[i]<=27:
            grade4+=1
        elif 28<=l1[i]+l2[i]<=30:
            grade5+=1
    return grade1,grade2,grade3,grade4,grade5,fail


pavg, passper, fail, grade1, grade2, grade3, grade4, grade5=analysis()
print("Statistics:")
print(f"Points average: {pavg:.1f}")
print(f"Pass percentage: {passper:.1f}")
print("Grade distribution:")
print(f"5: {'*'*grade5}")
print(f"4: {'*'*grade4}")
print(f"3: {'*'*grade3}")
print(f"2: {'*'*grade2}")
print(f"1: {'*'*grade1}")
print(f"0: {'*'*fail}")
````

````python
def exam_and_exercise_completed(inpt):
    space = inpt.find(" ")
    exam = int(inpt[:space])
    exercise = int(inpt[space+1:])
    return [exam, exercise]
 
def exercise_points(amount):
    return amount // 10
 
def grade(points):
    boundary = [0, 15, 18, 21, 24, 28]
    for i in range(5, -1, -1):
        if points >= boundary[i]:
            return i
 
def mean(points):
    return sum(points) / len(points)
 
def main():
    points = []
    grades = [0] * 6
    while True:
        inpt = input("Exam points and exercises completed: ")
        if len(inpt) == 0:
            break
 
        exam_and_exercises = exam_and_exercise_completed(inpt)
        exercise_pnts = exercise_points(exam_and_exercises[1])
        total_points = exam_and_exercises[0] + exercise_pnts
 
        points.append(total_points)
        grd = grade(total_points)
        if exam_and_exercises[0] < 10:
            grd = 0
        grades[grd] += 1
 
    pass_pros = 100 * (len(points) - grades[0]) / len(points)
 
    print("Statistics:")
    print(f"Points average: {mean(points):.1f}")
    print(f"Pass percentage: {pass_pros:.1f}")
    print("Grade distribution:")
    for i in range(5, -1, -1):
        stars = "*" * grades[i]
        print(f"  {i}: {stars}")
 
main()
 
````
