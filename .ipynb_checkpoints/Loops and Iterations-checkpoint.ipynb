# While Loops (Definite Iterations)

1. A while loop is a simple loop function used to run multiple iterations of a code.
2. Please note that only the code inside of a while loop is executed when it is in function. Also always ensure that there is a `break` condition for the loop else it will continue running ad infinitum.
3. Example code:

```python
while True:
    num=int(input("Please type in a number:"))
    if num==0:
        break
    print(num**2)
print("Thankyou for playing!")
```

4. To create a loop 3 main things are required: <br>
   i) Initialization: <br>
   This happens outside of the loop block and contains and defines all the variables used thereon along with any extra values.<br>
   ii) Loop block:<br>
   This contains the actual heart of loop, note that the condition of the `while` loop needn't always be true but can be any condition so long as it is syntactically valid.<br>
   iii) Post:<br>
   This is used to round up and print the output as required; a sample code to print the sum of numbers till an upper limit (The result  printed maybe more than the limit defined by the user is:

```python
lim=int(input("Limit:"))
series=1
add=1
while series<lim:
    add+=1
    series+=add
    
print(series)
```

Another example code working with strings and numbers to print out the sum (as in the previous code block) but also the calculation:

```python
lim=int(input("Limit:"))
series=1
add=1
asd=""
while series<lim:
    asd+=f"{add} + "
    add+=1
    series+=add
    
print(f"The consecutive sum: {asd} = {series}")
```

## Continue Command:
The Continue command is used to jump straight to the beggining of the loop and skip all conditions except the master condition of a loop


We can also nest loops inside other loops to simplify functioning, for example this block prints the multiplication tables till both numbers match a number defined by the user:

```python
num=int(input("Please type in a number:"))
i=1
while True:
    l=1
    while l<=num:
        print(f"{i} x {l} = {l*i}")
        l+=1
    if i==num:
        break
    i+=1
    
```
