Here are some things to keep in mind: 
1. Any condition used in a conditional statement will result in a ***truth value***, that is, either true or false. For example, the condition `a < 5` is true if `a` is less than 5, and false if `a` is equal to or greater than 5.

2. These types of values are often called _Boolean_ values, named after the English mathematician George Boole. In Python they are handled by the `bool` data type. Variables of type `bool` can only have two values: `True` or `False`.
3. The result of a boolean condition can be printed by using the statement `print(condition_name)`
4. The `len` function command can be used to return the length of a string.

5. Python comparisons can also be used on strings and alphabets, the < and > comparators compare which string comes alphabetically first.
6. `NOT` is a python operator used to negate a statement, it generally used to find if a statement if false, and if so it is used to to execute a statement, example: <br>
   `if not a>b: print("a is not greater than b)`
7. The `and`  and `or` conditionals are used to link two conditions, `and` incorporates the intersection of two sets, whereas `or` incorporates the addition of all variables of two sets.
8. In python a<=variable<=b is a shorthand way of writing that a variable is greater than `a` but smaller than `b`.

When using `if` conditions, ensure to first argue the bigger statement that incorporate multiple conditions `number%3==0 and number%5==0` incorporates two conditions, if however we were to place it after the `if` condition `number%3==0` then our larger statement would not be executed since an earlier statement has been executed and which resulted in the end of the `if block`

10. Nested conditionals means nesting `if` statements under other `if` statements, i.e. adding a condition inside a condition, however do not forget to use the correct indention in nested ifs and do not forget to write both true and false statements for each if. <br>
