# Dictionary:

Dictionaries are essentially bigger fish in the sea of storing information. <br>

Lists can be handy in many situations, but they are limited by the fact that the items are accessed through indexes; 0, 1, 2, and so forth. <br>If you want to find some item in a list, you will either have to know its index, or, at worst, traverse through the entire list.
<br> This is where dictionaries store information by corelating one piece of a variable (**value**) with a corresponding called the **key**. <br>


````python
my_dictionary = {}

my_dictionary["apina"] = "monkey"
my_dictionary["banaani"] = "banana"
my_dictionary["cembalo"] = "harpsichord"

print(len(my_dictionary))
print(my_dictionary)
print(my_dictionary["apina"])
````

The notation {} creates an empty dictionary, to which we can now add content. <br> Three key-value pairs are added:"apina" maps to "monkey", "banaani" maps to "banana", and "cembalo" maps to "harpsichord". <br> Finally, the number of key-value pairs in the dictionary is printed, along with the entire dictionary, and the value mapped to the key "apina".

In a similar manner, dictionary 'keys' can be anything from floats to ineteger to lists to even enitre dictionaries.
> Note: Even Though Dictionary keys can be lists, the values cannot be lists since the 'values' are said to be **hashable** i.e. each value of a dictionary is stripped of its properties and 'hashed' for the lack of better term and made into simple 1s and 0s which is immutable, however lists are inherently mutable and cannot be hashed like integers or floats, therefore they are not a valid type of 'value'.

````python
def factorials(integer):
    fact={}
    l=0
    fact[1]=1
    for i in range(2,integer+1):
        l=i
        for j in range(1,i):
            l=l*j
        fact[i]=l
    return fact

if __name__=="__main__":
    k = factorials(5)
    print(k[1])
    print(k[3])
    print(k[5])
````

Dictionaries like lists can be traversed by using `for loops` and the value, key pairs can be accessed as so: 
```python
for key, value in my_dictionary.items():
    print("key:", key)
    print("value:", value)
```
The method `items` returns all the keys and values stored in the dictionary, one pair at a time

A common use of dictionaries is to keep track of how many times a variable or a character has occured:
```python
def histogram(string):
    k={}
    for letter in string:
        if letter not in k:
            k[letter]=0
        k[letter]+=1
    
    for value, key in k.items():
        print(f"{value} {key*"*"}")

```
