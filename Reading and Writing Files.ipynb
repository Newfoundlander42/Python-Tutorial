```python
### Here is an example code that deals with a matrix
def open_matrix():
    with open("matrix.txt") as matrix:
        maitrix=[]
        for line in matrix:
            row=[]
            line=line.replace("\n", "")
            number_as_strings=line.split(",")
            for element in number_as_strings:
                row.append(int(element))
            maitrix.append(row)
    return maitrix

def matrix_sum():
    summation=0
    matrix=open_matrix()
    for row in matrix:
        summation+=sum(row)
    return summation

def matrix_max():
    maximus=[]
    matrix=open_matrix()
    for row in matrix:
        maximus.append(max(row))
    return max(maximus)

def row_sums():
    summation=[]
    matrix=open_matrix()
    for row in matrix:
        summation.append(sum(row))
    return summation

print(matrix_sum())
print(matrix_max())
print(row_sums())
```

```Python
### Here is a sample code that creates a dictionary based off of raw data

grades = {}
with open("grades.csv") as new_file:
    for line in new_file:
        line = line.replace("\n", "")
        parts = line.split(";")
        name = parts[0]
        grades[name] = []
        for grade in parts[1:]:
            grades[name].append(int(grade))

print(grades)
```
